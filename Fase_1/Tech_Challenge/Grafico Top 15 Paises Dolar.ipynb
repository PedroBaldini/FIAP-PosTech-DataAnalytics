```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Importando os Dados
df = pd.read_excel('DatasetFinal.xlsx')
```

```python
# Removendo as duas primeiras colunas que são desnecessárias
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# Verificando o tipo dos dados de cada coluna
df.dtypes
```

```python
# Realizando filtro para pegar apenas os últimos 15 anos para o tipo "Vinho Mesa"
df_filtrado = df[(df['Ano'] >= 2008) & (df['Tipo'] == 'Vinho Mesa')]
```

```python
# Realizando agrupamento dos dados para verificar os países com mais exportações nesse período vendo o total em dolar
vinho_mesa = df_filtrado.groupby(by=['Pais'])[['Valor_KG', 'Valor_Dolar']].sum().sort_values(by=['Valor_Dolar'], ascending = False)
vinho_mesa[:15]
```

```python
# Resetando o index e realizando a cópia do dataframe para a variável "vinho_mesa_paises_litros"
vinho_mesa_top15_dolar = vinho_mesa[:15].reset_index()

# Inserindo separador de milhar nas colunas numéricas
vinho_mesa_top15_dolar['Valor_KG'] = vinho_mesa_top15_dolar['Valor_KG'].map("{:,}".format)
vinho_mesa_top15_dolar['Valor_Dolar'] = vinho_mesa_top15_dolar['Valor_Dolar'].map("{:,}".format)

# Renomenado as colunas do dataframe
vinho_mesa_top15_dolar = vinho_mesa_top15_dolar.rename(columns={"Valor_KG": "Quantidade em litros de vinho exportado",
                                                                 "Valor_Dolar": "Valor total em dolar de vinho exportado"})
vinho_mesa_top15_dolar
```

```python
# Exportando para realizar análise no PBI
lista_paises_dolar = ["Russia", "Paraguai", "Estados Unidos da América", "China", "Reino Unido", "Espanha", 
                       "Holanda", "Alemanha" , "Japão", "Haiti", "Belgica", "Uruguai", "Canada", "Suica", "França"]

vinho_mesa_PBI = df[(df['Ano'] >= 2008) & (df['Tipo'] == 'Vinho Mesa') & (df['Pais'].isin(lista_paises_dolar))]
vinho_mesa_PBI.to_excel("TOP15_DOLARES.xlsx")
```
